```python
import os, shutil
import pickle
import subprocess
import numpy as np
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

TEMP_DIR = "temporary"
OUT_DIR = "output"
IMG_DIR = f"{OUT_DIR}/images"
FILE_PATH = "syscalls"
SYSCALLS = ["snd-cert", "snd-unm"]
LABELS = ["self","nonself"]

def create_files():
    for syscall in SYSCALLS:
        self_file = open(f"{TEMP_DIR}/{syscall}-self.test", "w")
        nonself_file = open(f"{TEMP_DIR}/{syscall}-nonself.test", "w")
        snd_path = f"{FILE_PATH}/{syscall}/{syscall}"
        for file_nr in range(1,4):
            with open(f"{snd_path}.{file_nr}.test") as test_file, open(f"{snd_path}.{file_nr}.labels") as labels_file:
                for syscall, lbl in zip(test_file, labels_file):
                    syscall = syscall.strip(); lbl = lbl.strip()
                    if int(lbl) == 1:
                        nonself_file.write(syscall+"\n")
                    else:
                        self_file.write(syscall+"\n")
        self_file.close()
        nonself_file.close()


def split_train(chunk_size):
    for syscall in SYSCALLS:
        syscalls_split_file = open(f"{TEMP_DIR}/{syscall}-split.train", "w")
        snd_train_path = f"{FILE_PATH}/{syscall}/{syscall}.train"
        with open(snd_train_path) as train_file:
            for syscall in train_file:
                n_g = extract_n_g(syscall.strip(),chunk_size,unique=True)
                if len(n_g)==0:
                    continue
                syscalls_split_file.writelines(n_g)
        syscalls_split_file.close()


def chunks(chunk_size):

    syscall_dict = dict()
    for syscall in SYSCALLS:
        for label in LABELS:
            syscall_positions = []
            syscall_pos = 0
            filename = f"{TEMP_DIR}/{syscall}-{label}"
            syscalls_split_file = open(filename+"-split.test", "w")
            with open(filename+".test") as syscalls_file:
                for syscall in syscalls_file:
                    n_g = extract_n_g(syscall.strip(),chunk_size,unique=False)
                    if len(n_g)==0:
                        continue
                    syscalls_split_file.writelines(n_g)
                    syscall_pos += len(n_g)
                    syscall_positions.append(syscall_pos)
            syscall_dict[f"{syscall}-{label}"] = syscall_positions
            syscalls_split_file.close()
    return syscall_dict


def extract_n_g(string, n, overlap=0, unique=True):
    if n > len(string):
        return []    

    n_g = []
    step = n - overlap
    n_g = [string[i:i+n] for i in range(0, len(string)-step+1, step)]

    if unique:
        n_g = list(set(n_g))

    n_g = [g+"\n" for g in n_g]

    return n_g


def merge_test_files():
    for syscall in SYSCALLS:
        self_file = open(f"{TEMP_DIR}/{syscall}-self-split.test")
        nonself_file = open(f"{TEMP_DIR}/{syscall}-nonself-split.test")
        merged_file = open(f"{TEMP_DIR}/{syscall}-merged-split.test", "w")
        merged_lines = self_file.readlines()
        merged_lines.extend(nonself_file.readlines())
        merged_file.writelines(merged_lines)
        self_file.close()
        nonself_file.close()
        merged_file.close()


def negative_selection(syscall, n, r, count=True, log=False):
    alphabet_file = f"file://{FILE_PATH}/{syscall}/{syscall}.alpha"
    self_file = f"{TEMP_DIR}/{syscall}-split.train"
    count_param = "-c" if count else ""
    log_param = "-l" if log else ""
    os.system(f"java -jar negsel2.jar -alphabet {alphabet_file} -self {self_file} -n {n} -r {r} \
                {count_param} {log_param} < {TEMP_DIR}/{syscall}-merged-split.test \
                > {OUT_DIR}/N={n}-R={r}-{syscall}-output")


def collect_pred(n, r, syscall, syscall_dict):
    output_file = open(f"{OUT_DIR}//N={n}-R={r}-{syscall}-output", "r")
    indexes = syscall_dict[f"{syscall}-self"]
    nonself_indexes = syscall_dict[f"{syscall}-nonself"]
    nonself_indexes = np.array(nonself_indexes) + indexes[-1]
    indexes.extend(nonself_indexes.tolist())
    predictions = []
    start = 0
    for i in indexes:
        results = [float(next(output_file).strip()) for _ in range(i-start)]
        total_anomaly = np.mean(results)
        predictions.append(total_anomaly)
        start = i

    return predictions

def ROC(n, r, ground, predictions):
    fpr, tpr, _ = roc_curve(ground, predictions)
    auc_score = auc(fpr, tpr)
    nas_fpr, nas_tpr, _ = roc_curve(ground, np.zeros(len(predictions),))
    plt.figure(figsize=(7, 7))
    plt.grid()
    plt.plot(fpr, tpr, marker='.', label=f'Negative Selection (AUC={auc_score:.3f})', linewidth=2)
    plt.plot(nas_fpr, nas_tpr, linestyle='--', label='Chance', linewidth=2)
    plt.xlabel('False Positive')
    plt.ylabel('True Positive')
    plt.title(f"ROC-curve for Negative Selection with N={n} and R={r}")
    plt.legend(loc=4, framealpha=1)

def ROC(n, r, ground, predictions):
    fpr, tpr, _ = roc_curve(ground, predictions)
    auc_score = auc(fpr, tpr)
    plt.figure(figsize=(7, 7))
    plt.grid()
    plt.plot(fpr, tpr, marker='.', label=f'Negative Selection (AUC={auc_score:.3f})', linewidth=2)
    plt.plot([0,1], [0,1], linestyle='--', label='No Skill', linewidth=2)
    plt.xlabel('False Positive')
    plt.ylabel('True Positive')
    plt.title(f"ROC-curve for Negative Selection with N={n} and R={r}")
    plt.legend(loc=4, framealpha=1)

def run_negative_selection(n, r):
    if not os.path.exists(TEMP_DIR):
        os.makedirs(TEMP_DIR)
    if not os.path.exists(OUT_DIR) or not os.path.exists(IMG_DIR):
        os.makedirs(IMG_DIR)

    print("Preparing files")
    create_files()
    syscall_dict = chunks(n)

    with open(f"{OUT_DIR}/N={n}-R={r}-indices.pkl", 'wb') as f:
        pickle.dump(syscall_dict, f, pickle.HIGHEST_PROTOCOL)
    split_train(n)
    merge_test_files()

    for syscall in SYSCALLS:
        print(f"Negative selection for {syscall}")
        negative_selection(syscall, n, r, count=True, log=True)

    print("Calculate results")
    for syscall in SYSCALLS:
        predictions = collect_pred(n, r, syscall, syscall_dict)
        plt.figure(figsize=(8,8))
        plt.hist(predictions)
        plt.savefig(f"{IMG_DIR}/N={n}-R={r}-{syscall}-hist.png")
        ground = np.zeros(len(predictions))
        num_of_nonself = len(syscall_dict[f"{syscall}-nonself"])
        ROC(n,r, ground, predictions)
        plt.savefig(f"{IMG_DIR}/N={n}-R={r}-{syscall}-roc.png")

    shutil.rmtree(TEMP_DIR)


def generate_overlapping_rocs(n_list, r_list):
    for syscall in SYSCALLS:
        plt.figure(figsize=(7, 7))
        plt.grid()
        plt.plot([0,1], [0,1], linestyle='--', label='No Skill', linewidth=2)
        for n,r in zip(n_list, r_list):
            with open(f'{OUT_DIR}/N={n}-R={r}-indices.pkl', 'rb') as f:
                syscall_dict = pickle.load(f)
            predictions = collect_pred(n, r, syscall, syscall_dict)
            ground = np.zeros(len(predictions))
            num_of_nonself = len(syscall_dict[f"{syscall}-nonself"])
            ground[-num_of_nonself:] = 1
            fpr, tpr, _ = roc_curve(ground, predictions)
            auc_score = auc(fpr, tpr)
            plt.plot(fpr, tpr, marker='.', label=f'N={n}, R={r} (AUC={auc_score:.3f})', linewidth=2)
        plt.xlabel('False Positive')
        plt.ylabel('True Positive')
        plt.title(f"ROC-curve on {syscall}")
        plt.legend(loc=3, framealpha=1)
        plt.savefig(f"{IMG_DIR}/{syscall}-roc.png")

def main():
    #n_list = [3,5,7]
    #r_list = [2,4,6]
    n_list = [7,7,7,7]
    r_list = [3,4,5,6]

    generate_overlapping_rocs(n_list, r_list)

if __name__ == "__main__":
    main()
```
